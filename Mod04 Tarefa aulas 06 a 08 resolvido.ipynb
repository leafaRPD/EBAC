````python
#1) Seu código aqui. Novamente, eu não sabia como colocar o 'uf' e o 'tipo_interessado' no meio do endereço. Daí, Chat.
import pandas as pd
import requests

def consultar_dados_tesouro(uf, tipo_interessado):
    url = f'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl?uf={uf}&tipo_interessado={tipo_interessado}'
    
    # Faz a requisição HTTP para a URL
    response = requests.get(url)
    
    # Verifica se a requisição foi bem-sucedida
    if response.status_code == 200:
        # Converte os dados JSON em um DataFrame usando o Pandas
        dados_json = response.json()
        df = pd.DataFrame(dados_json['items'])
        
        return df
    else:
        print(f"Erro ao consultar os dados. Código de status: {response.status_code}")
        return None

# Exemplo de uso da função
uf = 'AM'
tipo_interessado = 'Estado' # ou 'Munic%C3%ADpio'
df = consultar_dados_tesouro(uf, tipo_interessado)

# Imprime o DataFrame resultante
df
````
