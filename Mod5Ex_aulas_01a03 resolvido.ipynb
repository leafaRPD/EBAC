````python
import pandas as pd
import requests
# Carregue a base 'SINASC_RO_2019.csv'. Conte o número de registros e o número de registros não duplicados da base. 
# Dica: você aprendeu um método que remove duplicados, encadeie este método com um outro método que conta o número de linhas. 
# Há linhas duplicadas?

# 1) seu código aqui
sinasc = pd.read_csv('D:/Programação/EBAC/Curso Cientista de Dados - Módulo 5/Tarefa1/SINASC_RO_2019.csv')
print(sinasc.shape)
sinasc.drop_duplicates().shape
# Não há duplicados
````

````python
# Conte o número de valores *missing* por variável.
# 2) seu código aqui
sinasc.isna().sum()
````

````python
# Ok, no item anterior você deve ter achado pouco prático ler a informação de tantas variáveis, 
# muitas delas nem devem ser interesantes. Então crie uma seleção dessa base somente com as colunas que interessam. 
# São elas:
# ['LOCNASC', 'IDADEMAE', 'ESTCIVMAE', 'ESCMAE', 'QTDFILVIVO', 
#    'GESTACAO', 'GRAVIDEZ', 'CONSULTAS', 'APGAR5'] 
#
# Refaça a contagem de valores *missings*.  
# 3) seu código aqui
colunas_desejadas = ['LOCNASC', 'IDADEMAE', 'ESTCIVMAE', 'ESCMAE', 'QTDFILVIVO', 'GESTACAO', 'GRAVIDEZ', 'CONSULTAS', 'APGAR5']
sinasc_interesse = sinasc[colunas_desejadas]

valores_ausentes = sinasc_interesse.isna().sum()
valores_ausentes
````

````python
# Apgar é uma nota que o pediatra dá ao bebê quando nasce de acordo com algumas características associadas principalmente à 
# respiração. Apgar 1 e Apgar 5 são as notas 1 e 5 minutos do nascimento. Apgar5 será a nossa variável de interesse principal.
# Então remova todos os registros com Apgar5 não preenchido.
# Para esta seleção, conte novamente o número de linhas e o número de missings.
# 4) seu código aqui
sinasc_interesse = sinasc_interesse.dropna(subset=['APGAR5'])
sinasc_interesse
````

````python
valores_ausentes = sinasc_interesse.isna().sum()
valores_ausentes
````

````python
# observe que as variáveis ['ESTCIVMAE', 'CONSULTAS'] possuem o código 9, que significa ignorado. 
# Vamos assumir que o não preenchido é o mesmo que o código 9.
# 5) seu código aqui
sinasc_interesse['ESTCIVMAE'] = sinasc_interesse['ESTCIVMAE'].fillna('9')
sinasc_interesse['CONSULTAS'] = sinasc_interesse['CONSULTAS'].fillna('9')
sinasc_interesse

````

````python
# Substitua os valores faltantes da quantitativa (QTDFILVIVO) por zero.
# 6) Seu código aqui
sinasc_interesse['QTDFILVIVO'] = sinasc_interesse['QTDFILVIVO'].fillna('0')
sinasc_interesse
````

````python
# Das restantes, decida que valore te parece mais adequado (um 'não preenchido' ou um valor 'mais provável' como no item 
# anterior) e preencha. Justifique. Lembre-se de que tratamento de dados é trabalho do cientista, e que estamos tomando 
# decisões a todo o momento - não há necessariamente certo e errado aqui.
# 7) seu código aqui
# colunas a serem tratadas: ['LOCNASC', 'IDADEMAE', 'ESCMAE', 'GESTACAO', 'GRAVIDEZ']
# Decidiu-se tratá-las como '9', ou seja, ignorada. Apenas a 'idade mãe', que não pôde ser tratada como 'ignorada' por
# não haver um código para isso, que será editada de modo diferente:
# Os valores faltantes serão trocados pela mediana.

sinasc_interesse['LOCNASC'] = sinasc_interesse['LOCNASC'].fillna('9')
sinasc_interesse['IDADEMAE'] = sinasc_interesse['IDADEMAE'].fillna(sinasc_interesse['IDADEMAE'].median())
sinasc_interesse['ESCMAE'] = sinasc_interesse['ESCMAE'].fillna('9')
sinasc_interesse['GESTACAO'] = sinasc_interesse['GESTACAO'].fillna('9')
sinasc_interesse['GRAVIDEZ'] = sinasc_interesse['GRAVIDEZ'].fillna('9')
sinasc_interesse
````

````python
sinasc_interesse['IDADEMAE'].median()
````

````python
valores_ausentes = sinasc_interesse.isna().sum()
valores_ausentes
````

````python
# O Apgar possui uma classificação indicando se o bebê passou por asfixia:
# 
# Entre 8 e 10 está em uma faixa 'normal'.
# Entre 6 e 7, significa que o recém-nascido passou por 'asfixia leve'.
# Entre 4 e 5 significa 'asfixia moderada'.
# Entre 0 e 3 significa 'asfixia severa'.
# Crie uma categorização dessa variável com essa codificação e calcule as frequências dessa categorização.

# 8) seu código aqui
sinasc_interesse['APGAR5'].value_counts()
````

````python
sinasc_interesse.loc[sinasc_interesse['APGAR5'] >7, 'Apgar5-cat'] = 'normal'
sinasc_interesse.loc[(sinasc_interesse['APGAR5'] >5) & (sinasc_interesse['APGAR5']<=7), 'Apgar5-cat'] = 'asfixia leve'
sinasc_interesse.loc[(sinasc_interesse['APGAR5'] >3) & (sinasc_interesse['APGAR5']<=5), 'Apgar5-cat'] = 'asfixia moderada'
sinasc_interesse.loc[sinasc_interesse['APGAR5'] <=3, 'Apgar5-cat'] = 'asfixia severa'
sinasc_interesse.head(21)
````

````python
# Renomeie as variáveis para que fiquem no snake case, ou seja, em letras minúsculas, com um underscore entre as palávras. 
# Dica: repare que se você não quiser criar um dataframe novo, você vai precisar usar a opção inplace = True.
# 9) seu código aqui
colunas_atuais = ['LOC NASC', 'IDADE MAE', 'EST CIV MAE', 'ESC MAE', 'QTD FIL VIVO', 'GESTACAO', 'GRAVIDEZ', 'CONSULTAS', 'APGAR5', 'apgar5_cat']

colunas_corrigidas = {}
for coluna in colunas_atuais:
    colunas_corrigidas[coluna] = coluna.lower().replace(" ", "_")

sinasc_interesse.rename(columns=colunas_corrigidas, inplace=True)
sinasc_interesse
````
